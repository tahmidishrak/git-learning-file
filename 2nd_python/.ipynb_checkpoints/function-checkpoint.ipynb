```python
def calcu_profit_cost (reve, cost): #always remember name spell
    profit = reve - cost
    margin = (reve / cost) * 100
    print(f"Profit : {profit:.2f}")
    print(f"Margin : {margin:.2f}")
```

```python
reve = 1000
cost = 700

calcu_profit_cost(reve , cost) #always remember name spell
```

```python
def avarage(*revenue):
    total = 0 
    for i in revenue:
        total += i      #hare has 2 way to right this total = total + i
    print("Average :"
          , total / len(revenue))
```

```python
avarage(40, 49, 50, 20, 29)
```

```python
def cha_bill(cup, *addons, **discount):
    base_price = cup * 15
    addons_price = len(addons) * 5
    sub_total = base_price + addons_price
    discount_percent = discount.get("discount_percent", 0)   # safe default 0
    total_bill = sub_total - (sub_total * discount_percent / 100)
    
    print(f"Cups ordered : {cup} and total Cups price {base_price}")
    if addons:
        print(f"Addons ordered : {addons} and total Addons price {addons_price}")
    print(f"Subtotal: {sub_total}")
    if "discount_percent" in discount:
        print(f"Discount applied: {discount_percent}%")
    print(f"Total bill: {total_bill}")

    return total_bill
```

```python
cha_bill(3)
```
