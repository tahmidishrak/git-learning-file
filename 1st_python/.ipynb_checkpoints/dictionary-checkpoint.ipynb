```python
#Calculate profit margin for each revenue-cost pair.
#Margin = (Profit / Revenue) * 100
```

```python
revenues = [20, 90, 80, 30]
cost = [83, 83, 90, 94]
```

```python
for i in range (len(revenues)):
    profit = revenues[i] - cost[i]
    margin = profit / revenues[i] * 100
    print(f"Margin Q.{i+1} : {margin: .2f}")
```

```python
#create a list of margin
```

```python
margins = []
for i in range (len(revenues)):
    profit = revenues[i] - cost[i]
    margin = profit / revenues[i] * 100
    margins.append(margin)

margins
```

```python
financial = [
    {"revenue": 70, "cost": 60},
    {"revenue": 80, "cost": 50},
    {"revenue": 90, "cost": 40}
]
```

```python
for i in financial:
    profit = i["revenue"] - i["cost"]
    i["margin"] = profit / i["revenue"] * 100
financial
```

```python
#python nested dictionary
```

```python
financial = {
    "Q1" :{"revenue": 50, "cost": 10},
    "Q2" :{"revenue": 60, "cost": 30},
    "Q3" :{"revenue": 55, "cost": 35},
    "Q4" :{"revenue": 10, "cost": 35},
}
```

```python
for quarter, data in financial.items():
    print("Quarter :",quarter)
    print("Data :" ,data)
```

```python
# python nested dictionary function
```

```python
for quarter, data in financial.items():
    margin = (data["revenue"] - data["cost"]) / data["revenue"] * 100
    print(f"{quarter} : {margin:.2f}% ")
```

```python
#python nested dictionary function with if stetment
```

```python
for quarter, data in financial.items():
    if data["revenue"] < 0:
        print(f"Invalied revnue for {quarter}. Skipping")
        continue
    margin = (data["revenue"] - data["cost"]) / data["revenue"] * 100
    print(f"{quarter} : {margin:.2f}% ")  
```

```python
students = {
    "Aarav":   [85, 90, 78],
    "Priya":   [72, 68, 75],
    "Rohan":   [45, 52, 48],
    "Sneha":   [95, 92, 98],
    "Manish":  [60, 65, 70],
    "tahmid": [30, 20, 39],
}
```

```python
for name, mark in students.items():
    avarage = sum(mark) / len(mark)
    if avarage >= 80:
        print(f"{name} your mark : {avarage:.2f}")
        print(f"{name} you are Topper")
        continue
    if  60 <= avarage <= 70:
        print(f"{name} your mark : {avarage:.2f}")
        print(f"{name} you are Avarage")
        continue
    if 33 <= avarage < 60:
        print(f"{name} your mark : {avarage:.2f}")
        print(f"{name} you are Pass")
        continue
    if avarage < 33:
        print(f"{name} your mark : {avarage:.2f}")
        print(f"{name} you need improve you are fail")
        continue
    
```
