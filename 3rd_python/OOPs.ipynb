```python
class car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def full_name(self):
        return (f"{self.brand} {self.model}")

my_car = car("Toyota" ,"skyline GTr34")
print(my_car.brand)
print(my_car.model)
print(my_car.full_name())
```

```python
class car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def full_name(self):
        return (f"{self.brand} {self.model}")

class electric_car(car):
    def __init__ (self, brand, model, battery_size):
        super().__init__(brand, model)
        self.battery_size = battery_size

my_electric_car = electric_car("tesla", "s", "80kWh")
print(my_electric_car.full_name())
```

```python
class YouTubeChannel:
    def __init__(self, name):
        self.name = name
        self.subscribers = 0   # শুরুতে 0 subscriber

    def subscribe(self):
        self.subscribers += 1
        print(f"Thanks for subscribing! Total subscribers: {self.subscribers}")

    def unsubscribe(self):
        if self.subscribers > 0:   # কখনো 0-এর নিচে যাবে না
            self.subscribers -= 1
        print(f"Total subscribers: {self.subscribers}")

    def show(self):
        print(f"{self.name} has {self.subscribers} subscribers")
```

```python
# Object তৈরি করা
user = YouTubeChannel("Tahmid")

# Test run
user.subscribe()
user.subscribe()
user.subscribe()
user.unsubscribe()
user.show()
```
